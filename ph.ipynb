```python

ph = int(input("dime el PH del elemento"))

if ph < 4:
  print("Es Muy ácido")
elif 4 <= ph < 7:
  print("Es Poco ácido")
elif ph == 7:
  print("Es neutro")
elif 7 < ph < 10:
  print("Es Poco Alcalino")
else:
  print("Es Muy Alcalino")
```
