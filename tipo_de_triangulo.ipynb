#**Tipo de triÃĄngulo: acutÃĄngulo, obtusÃĄngulo o rectÃĄngulo**
Conociendo la longitud sus lados, podemos averiguar quÃĐ tipo de triÃĄngulo tenemos:


1.   $ladoMenor^2 + ladoMediano^2 = ladoMayor^2$ $->$ **Es un triÃĄngulo rectÃĄngulo**
2.   $ladoMenor^2 + ladoMediano^2 < ladoMayor^2$ $->$ **Es un triÃĄngulo obtusÃĄngulo**
1.   $ladoMenor^2 + ladoMediano^2 > ladoMayor^2$ $->$ **Es un triÃĄngulo acutÃĄngulo**

Usando sentencias if_elif_else debemos determinar quÃĐ tipo de triÃĄngulo tenemos

```python
#Introducimos los valores de los lados

ladoMenor = int(input("dime el valor del lado Menor: "))
ladoMediano = int(input("dime el valor del lado Mediano: "))
ladoMayor = int(input("dime el valor del lado Mayor: "))

#implementamos las declaraciones if_elif_else donde se imprima el tipo de triÃĄngulo

if ladoMenor**2 + ladoMediano**2 == ladoMayor**2:
  print("Es un triÃĄngulo rectÃĄngulo")
elif ladoMenor**2 + ladoMediano**2 < ladoMayor**2:
  print("Es un triÃĄngulo obtusÃĄngulo")
else:
  print("Es un triÃĄngulo acutÃĄngulo")

```
